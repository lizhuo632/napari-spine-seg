```python

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageOps

path = '/share/data/CryoET_Data/lizhuo/SIM-data/actin_train_data/all/train_data/1.tif'
image = Image.open(path)
```

```python
plt.imshow(image)
```

```python
np.max(np.array(image))
```

```python
max_percentile = np.percentile(image, 99.95)
image = np.minimum(image, max_percentile)
image_pil = Image.fromarray(image).convert('L')

image = np.array(image)
image = image.astype(np.float32)
image = (image - np.min(image)) / (np.max(image) - np.min(image))
image = image * 65535
image = image.astype(np.uint16)

plt.imshow(image)
print(np.max(np.array(image)))
```

```python
image = ImageOps.equalize(image_pil)

plt.imshow(image)
```
